#### **requests 모듈**
1. http의 요청(rquest), 응답(response)을 위한 모듈이다. 

```python
import requests
```

#### get 요청하기
1. http get방식으로 요청한다.
2. 쿼리문자열을 이용하여 데이터를 전달한다.

```python
url='https://news.v.daum.net/v/20201230100440387'

resp = requests.get(url)
resp.status_code
```

```python
resp.text
```

#### post 요청하기
1. http post방식으로 요청한다.
2. http  body를 이용하여 데이터 전달한다.

```python
url = 'https://www.kangcom.com/member/member_check.asp'
data = {
    'id':'carami',
    'pwd':'TTty3@x!AX3R'
}

resp = requests.post(url,data=data)
resp.text
```

#### HTTP header 데이터 이용하기
1. header 데이터 구성하기
2. header 데이터 전달하기

```python
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
}
url='https://news.v.daum.net/v/20201230100440387'
resp = requests.get(url, headers=headers)
resp.text
```

#### HTTP response 처리하기
1. response 객체의 이해
2. status_code 확인하기
3. text 속성 확인하기

```python
url='https://news.v.daum.net/v/20201230100440387'
resp = requests.get(url)
if resp.status_code == 200:
    print(resp.text)
else :
    print('error')
```
